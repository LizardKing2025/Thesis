```python
from copy import deepcopy as dc

def prepare_dataframe(df, n_steps):
    df = dc(Datos)

    df.set_index('Fecha', inplace=True)

    for i in range(1, n_steps+1):
        Datos[f'Q_Ventas(t-{i})'] = Datos['Q_Ventas'].shift(i)

    df.dropna(inplace=True)

    return df

```

Convierto los subsets en estructura de serie de tiempo para usarlo en pytorch, esto se hace para mejorar eficiencia

```python
from torch.utils.data import Dataset

class TimeSeriesDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, i):
        return self.X[i], self.y[i]

train_dataset = TimeSeriesDataset(X_train, y_train)
test_dataset = TimeSeriesDataset(X_test, y_test)
```

Creo el modelo,

```python
class LSTM(nn.Module): # class hereda de nn.module
    def __init__(self, input_size, hidden_size, num_stacked_layers):
        super().__init__() # constructor de clase
        self.hidden_size = hidden_size #Guarda el tamaño de las capas ocultas en el atributo de instancia hidden_size.
        self.num_stacked_layers = num_stacked_layers # Guarda el número de capas apiladas en el atributo de instancia num_stacked_layers.

        self.lstm = nn.LSTM(input_size, hidden_size, num_stacked_layers,
                            batch_first=True) # Define una capa LSTM con el tamaño de entrada (input_size), tamaño de las capas ocultas (hidden_size), y el número de capas apiladas (num_stacked_layers). batch_first=True indica que la dimensión del batch es la primera dimensión en la entrada.

        self.fc = nn.Linear(hidden_size, 1) #Define una capa totalmente conectada (linear) que toma como entrada el tamaño de la capa oculta y produce una salida de tamaño 1

    def forward(self, x): # Define el método forward que especifica cómo los datos pasan a través de la red durante el paso hacia adelante (forward pass).
        batch_size = x.size(0) # Obtiene el tamaño del batch de la entrada x
        h0 = torch.zeros(self.num_stacked_layers, batch_size, self.hidden_size).to(device) # Inicializa el estado oculto h0 con ceros. Su tamaño es (num_stacked_layers, batch_size, hidden_size). to(device) asegura que los tensores estén en el mismo dispositivo (CPU o GPU) que los datos de entrada.
        c0 = torch.zeros(self.num_stacked_layers, batch_size, self.hidden_size).to(device) # Inicializa la celda de memoria c0 con ceros, de tamaño similar al estado oculto.

        out, _ = self.lstm(x, (h0, c0)) # Pasa la entrada x y los estados iniciales (h0, c0) a través de la capa LSTM. La salida out contiene las secuencias de salida de la LSTM para cada paso de tiempo.
        out = self.fc(out[:, -1, :]) # Toma la salida del último paso de tiempo (out[:, -1, :]) y la pasa a través de la capa totalmente conectada para producir la salida final.
        return out #Devuelve la salida de la red

model = LSTM(1, 4, 1) #Crea una instancia del modelo LSTM con un tamaño de entrada de 1, un tamaño de capa oculta de 4 y 1 capa apilada
model.to(device)
model

# En el contexto de redes neuronales recurrentes (RNNs) y específicamente en LSTM (Long Short-Term Memory), una "capa apilada" (stacked layer) se refiere a la práctica de tener múltiples capas de LSTM apiladas una encima de la otra. Esto significa que la salida de una capa LSTM se convierte en la entrada de la siguiente capa LSTM.
```

```python
# Agregar listas para almacenar la pérdida
train_losses = []
val_losses = []
```

Entrenamiento...

```python
def train_one_epoch():
    model.train(True)
    print(f'Epoch: {epoch + 1}')
    running_loss = 0.0 # Inicializa una variable para acumular la pérdida a lo largo de los lotes en la época.

    for batch_index, batch in enumerate(train_loader): # Itera sobre los lotes de datos proporcionados por el train_loader. enumerate proporciona tanto el índice del lote (batch_index) como el lote en sí (batch).
        x_batch, y_batch = batch[0].to(device), batch[1].to(device) # Extrae las características (x_batch) y las etiquetas (y_batch) del lote y las mueve al dispositivo especificado (CPU o GPU).

        output = model(x_batch) # Realiza el paso hacia adelante (forward pass) a través del modelo con las características del lote, produciendo las predicciones (output).
        loss = loss_function(output, y_batch) # Calcula la pérdida entre las predicciones del modelo y las etiquetas verdaderas utilizando la función de pérdida definida (loss_function).
        running_loss += loss.item() # Acumula la pérdida del lote actual en running_loss

        optimizer.zero_grad() # Resetea los gradientes de los parámetros del modelo. Esto es necesario porque por defecto, PyTorch acumula los gradientes.
        loss.backward() # Calcula los gradientes de la pérdida respecto a los parámetros del modelo (backward pass).
        optimizer.step() # Actualiza los parámetros del modelo utilizando los gradientes calculados.

        # Calcular la pérdida promedio del epoch
        epoch_loss = running_loss / len(train_loader)
        train_losses.append(epoch_loss)


        if batch_index % 100 == 99:  # print every 100 batches
            avg_loss_across_batches = running_loss / 100 #Comprueba si el índice del lote es un múltiplo de 100 (menos 1 debido a que batch_index comienza en 0). Esto asegura que se imprime la pérdida cada 100 lotes.

            print('Batch {0}, Loss: {1:.3f}'.format(batch_index+1,
                                                    avg_loss_across_batches))
            running_loss = 0.0
    print()
```

Validación...

```python
def validate_one_epoch():
    model.train(False) # Pone el modelo en modo de evaluación (en contraste con el modo de entrenamiento que se usa con model.train(True)). Esto asegura que capas como Dropout y BatchNorm se comporten de manera determinista durante la evaluación.
    running_loss = 0.0

    for batch_index, batch in enumerate(test_loader):
        x_batch, y_batch = batch[0].to(device), batch[1].to(device)

        with torch.no_grad(): # Temporalmente deshabilita el cálculo y almacenamiento de gradientes en PyTorch para reducir el uso de memoria y acelerar el cálculo. Esto es útil durante la evaluación cuando no necesitamos ajustar los pesos del modelo.
            output = model(x_batch)
            loss = loss_function(output, y_batch)
            running_loss += loss.item()

    # Calcular la pérdida promedio del epoch
    epoch_loss = running_loss / len(test_loader)
    val_losses.append(epoch_loss)

    avg_loss_across_batches = running_loss / len(test_loader)
    val_losses.append(epoch_loss)

    print('Val Loss: {0:.3f}'.format(avg_loss_across_batches))
    print('***************************************************')
    print()
```

```python
learning_rate = 0.001
num_epochs = 100
loss_function = nn.MSELoss() #  calcula el error cuadrático medio entre las predicciones y las etiquetas verdaderas.
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

for epoch in range(num_epochs):
    train_one_epoch()
    validate_one_epoch()
```
